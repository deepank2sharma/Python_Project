```python
# importing packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# importing dataset 

data = pd.read_csv("C:/Users/Deepank/Desktop/Simplylearn_Python_Case Study/World Hapiness Case Study/world-happiness-report-2021.csv")
```

```python
data.head(1)
```

```python
# Renaming Ladder score column to Happiness score

data.rename(columns={"Ladder score":"Happiness score"},inplace=True)
```

```python
data.head()
```

```python
# columns of our intrest

data_columns = ["Country name","Regional indicator","Happiness score","Logged GDP per capita","Social support","Healthy life expectancy","Freedom to make life choices","Generosity","Perceptions of corruption"]
```

```python
# subsetting the data

data = data[data_columns].copy()
```

```python
data.head()
```

```python
# Renaming all column names by replacing space with _


data.rename(columns={'Country name':'Country_name', 'Regional indicator':'Regional_indicator',"Ladder score":"Happiness_score"
                     ,'Happiness score':'Happiness_score','Logged GDP per capita':'Logged_GDP_per_capita'
                     , 'Social support':"Social_support", 'Healthy life expectancy':'Healthy_life_expectancy'
                     ,'Freedom to make life choices':'Freedom_to_make_life_choices', 'Generosity':'Generosity'
                     ,'Perceptions of corruption':'Perceptions_of_corruption'}, inplace=True)
```

```python
happy_df = data
```

```python
happy_df.head()
```

```python
# checking for null values

happy_df.isnull().sum()
```

```python
# creating plot b/w happiness score vs GDP for different region

plt.figure(figsize=(20,10))
plt.title("happiness score vs GDP",size=30)
sns.scatterplot(x=happy_df.Happiness_score,y=happy_df.Logged_GDP_per_capita, hue = happy_df.Regional_indicator, s= 150)
plt.xlabel("happiness score",size=20)
plt.ylabel("GDP",size=20)
plt.grid("True")
plt.show()
```

```python
# Creating pie chart of GDP for each region
```

```python
gdp_region = happy_df.pivot_table(index="Regional_indicator", values = "Logged_GDP_per_capita", aggfunc="sum").reset_index()
```

```python
gdp_region
```

```python
plt.figure(figsize=(15,7))
plt.title("GDP by region",size=20)
plt.pie(gdp_region.Logged_GDP_per_capita, labels=gdp_region.Regional_indicator, autopct='%1.1f%%', startangle=150, shadow=True);
plt.show()
```

```python
# Finding total number countries in each region

total_country = happy_df.pivot_table(index=["Regional_indicator"], values="Country_name", aggfunc="count").reset_index()
total_country
```

```python
# finding correlation

correlation = happy_df.corr()

correlation
```

```python
# Creating Heatmap

plt.figure(figsize=(10,5))
sns.heatmap(correlation, annot=True)
```

```python
# Creating bar plot that will show corruption in different regions

corruption = happy_df.pivot_table(index="Regional_indicator", values="Perceptions_of_corruption", aggfunc="mean").reset_index()
```

```python
corruption
```

```python
plt.figure(figsize=(15,7))
plt.title("corruption in different regions",size=20)

sns.barplot(x=corruption.Regional_indicator, y=corruption.Perceptions_of_corruption)

plt.xlabel("Region",size=15)
plt.xticks(rotation = 30, ha="right")
plt.ylabel("Perceptions_of_corruption",size=15)

plt.show()
```

```python
# Life Expactancy of Top 10/bottom 10 happy countries

top_10 = happy_df.loc[:,["Country_name","Happiness_score","Healthy_life_expectancy"]].sort_values(by="Happiness_score",ascending=False).head(10)
```

```python
top_10
```

```python
bottom_10 = happy_df.loc[:,["Country_name","Happiness_score","Healthy_life_expectancy"]].sort_values(by="Happiness_score",ascending=True).head(10)
```

```python
bottom_10
```

```python
combined = top_10.append(bottom_10)   ## appending top_10 & bottom_10
```

```python
combined
```

```python
combined.loc[:,"Top_Bottom_10"] = pd.cut(combined.Healthy_life_expectancy, bins=2, labels=["Bottom_10","Top_10"])  ##creating bins
```

```python
combined
```

```python
# Bar plot showing Life Expectancy for Top/Bottom 10 Happiest Country

plt.figure(figsize=(25,10))
plt.title("Life Expectancy for Top/Bottom 10 Happiest Country", size = 30)
sns.barplot(x=combined.Country_name, y=combined.Healthy_life_expectancy,hue=combined.Top_Bottom_10)
plt.xlabel("Country_name",size = 25)
plt.ylabel("Healthy_life_expectancy",size = 25)
plt.xticks(rotation = 30, ha="right",size=20)
plt.yticks(size=20)
plt.legend(fontsize="16")
plt.show()
```

```python
# Creating plot b/w Freedom to make life choices & Happiness score

plt.figure(figsize=(20,10))
plt.title("Freedom to make life choices vs Happiness", size = 30)
sns.scatterplot(x=happy_df.Happiness_score, y=happy_df.Freedom_to_make_life_choices,hue=happy_df.Regional_indicator,s=200)
plt.xlabel("Happiness_score",size = 25)
plt.ylabel("Freedom_to_make_life_choices",size = 25)
plt.xticks(size=20)
plt.yticks(size=20)
plt.legend(fontsize="16")
plt.show()
```

```python
# Finding to 10 most corrupt countries

top_corrupt_countries = happy_df.loc[:,["Country_name","Perceptions_of_corruption"]].sort_values(by="Perceptions_of_corruption", ascending=False).head(10).reset_index()
```

```python
top_corrupt_countries
```

```python
plt.figure(figsize=(10,5))
plt.title("10 most corrupt countries", size = 20)
sns.barplot(x=top_corrupt_countries.Country_name , y=top_corrupt_countries.Perceptions_of_corruption)
plt.xlabel("Country_name",size = 15)
plt.ylabel("Perceptions_of_corruption",size = 15)
plt.xticks(size=10)
plt.yticks(size=10)
plt.xticks(rotation = 30, ha="right")
plt.show()
```

```python
# Creating scatter plot for Corruption vs Happiness Score

plt.figure(figsize=(20,10))
plt.title("Corruption vs Happiness Score", size = 30)
sns.scatterplot(x=happy_df.Happiness_score, y=happy_df.Perceptions_of_corruption,hue=happy_df.Regional_indicator,s=200)
plt.xlabel("Happiness_score",size = 25)
plt.ylabel("Perceptions_of_corruption",size = 25)
plt.xticks(size=20)
plt.yticks(size=20)
plt.legend(fontsize="16")
plt.show()
```
